```python
hitlist = pd.concat([tracks['track'][['listens']], features['mfcc']], axis=1)
hitlist_temp = pd.concat([tracks['track'][['genre_top','listens']], features['mfcc']], axis=1)
grouped_hitlist = hitlist_temp.groupby('genre_top')
grouped_hitlist.get_group('Rock')
hitlist
```

```python
# a = hitlist['listens'].unique()
print(hitlist['listens'].quantile(0.9))
hitlist['listens'] = np.where(hitlist['listens'] >= 4866, "Yes", "No")
hitlist
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import plot_tree, plot_importance

from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
```

```python
Y=hitlist.loc[:,'listens']

X=hitlist.loc[:,hitlist.columns!='listens']
X.columns = [x[1] for x in X.columns]

cols=X.columns

scaler=preprocessing.MinMaxScaler()
scaled_X=scaler.fit_transform(X)

X=pd.DataFrame(scaled_X,columns=cols)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
```

```python
def model_assess(model,title):
    model.fit(X_train,y_train)
    preds=model.predict(X_test)
    print('Accuracy', title, ':', round(accuracy_score(y_test, preds), 5), '\n')
```

```python
# Naive Bayes
nb = GaussianNB()
model_assess(nb, "Naive Bayes")

# KNN
knn = KNeighborsClassifier(n_neighbors=19)
model_assess(knn, "KNN")

# Decision trees
tree = DecisionTreeClassifier()
model_assess(tree, "Decision trees")

# Random Forest
rforest = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=0)
model_assess(rforest, "Random Forest")

# Support Vector Machine
svm = SVC(decision_function_shape="ovo")
model_assess(svm, "Support Vector Machine")

# Logistic Regression
lg = LogisticRegression()
model_assess(lg, "Logistic Regression")
```
